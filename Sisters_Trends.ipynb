```python
df.columns=["month","Kim","Khole","Kourtney","Kendall","Kylie"]
```

```python
for column in df.columns:
    if "<" in df[column].to_string():
        df[column]=df[column].str.replace("<","")
        df[column]=pd.to_numeric(df[column])
```
